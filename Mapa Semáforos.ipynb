```python
import pandas as pd
import folium
from google.colab import files


df = pd.read_csv("Red_Semaf%C3%B3rica_de_Bogot%C3%A1_D.C..csv")


print(df.columns.tolist())
print(df.head())
```

```python
import pandas as pd
import folium

print(df[['X', 'Y']].head(10))
```

```python
import folium


df = df.dropna(subset=['X', 'Y'])


mapa = folium.Map(location=[4.7110, -74.0721], zoom_start=12, tiles='OpenStreetMap')

for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Y'], row['X']],
        radius=3,
        color='red',
        fill=True,
        fill_opacity=0.6,
        popup=folium.Popup(
            f"<b>Dirección:</b> {row['DIRECCION']}<br>"
            f"<b>Localidad:</b> {row['LOCALIDAD']}<br>"
            f"<b>Funcionamiento:</b> {row['FUNCIONAMIENTO']}",
            max_width=200
        )
    ).add_to(mapa)


mapa
```

```python
mapa.save("semaforos_bogota.html")
files.download("semaforos_bogota.html")
```
